```python
import cv2

cam = cv2.VideoCapture(0)

cv2.namedWindow("test")

img_counter = 0

while True:
    ret, frame = cam.read()
    if not ret:
        print("failed to grab frame")
        break
    cv2.imshow("test", frame)

    k = cv2.waitKey(1)
    if k%256 == 27:
        # ESC pressed
        print("Escape hit, closing...")
        break
    elif k%256 == 32:
        # SPACE pressed
        img_name = "opencv_frame_1.png".format(img_counter)
        cv2.imwrite(img_name, frame)
        print("{} written!".format(img_name))
        img_counter += 1

cam.release()

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('opencv_frame_1.png' )
img_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb);
```

```python
print('image dtype ',img_rgb.dtype) 
```

```python
blank_img = np.zeros(shape=(img_rgb.shape[0],img_rgb.shape[1],3), dtype=np.uint8)  
font = cv2.FONT_HERSHEY_SIMPLEX  
cv2.putText(blank_img,  
            text='justine',  
            org=(img_rgb.shape[0]//8, img_rgb.shape[0]//8),   
            fontFace=font,  
            fontScale= 1,color=(600,400,0),  
            thickness=1,  
            lineType=cv2.LINE_4)  
plt.imshow(blank_img);  
```

```python
blended = cv2.addWeighted(src1=img_rgb,alpha=0.9,src2=blank_img,beta=1, gamma = 0)  
plt.imshow(blended);
```

```python
cv2.imwrite('opencv_frame_{}.jpg', cv2.cvtColor(blended, cv2.COLOR_RGB2BGR)) 
```
